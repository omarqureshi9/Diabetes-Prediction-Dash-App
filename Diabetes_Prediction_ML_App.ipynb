```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/diabetes_data_upload.csv')
dataset
```

```python
data=dataset.copy()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
df_dummies = pd.get_dummies(data, columns=['Gender', 'Polyuria', 'Polydipsia', 'sudden weight loss',
       'weakness', 'Polyphagia', 'Genital thrush', 'visual blurring',
       'Itching', 'Irritability', 'delayed healing', 'partial paresis',
       'muscle stiffness', 'Alopecia', 'Obesity', 'class'], drop_first=True)
```

```python
df_dummies=df_dummies.astype(int)
```

```python
#Creating dummies in form of 1,0
df_dummies
```

```python
#Separating data into input and output variable and checking correlation
data=df_dummies
X=data.drop(columns=['class_Positive'])
y=data[['class_Positive']]
X.corr()
```

```python
#Dropping columns with high correlation
X=X.drop(columns=['Polydipsia_Yes','Polyuria_Yes','visual blurring_Yes','Itching_Yes'])
```

```python
#Standardizing X variables
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
X = pd.DataFrame(X, columns=['Age','Gender_Male','sudden weight loss_Yes','weakness_Yes','Polyphagia_Yes','Genital thrush_Yes','Irritability_Yes','delayed healing_Yes','partial paresis_Yes','muscle stiffness_Yes','Alopecia_Yes','Obesity_Yes'])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size =0.7)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

model = LogisticRegression()
model.fit(X_train, y_train)

# Making predictions
y_pred = model.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculating confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 5  # Number of neighbors
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from io import BytesIO
import base64

def train_model(X_train, y_train):
    model = DecisionTreeClassifier(criterion='entropy', random_state=0)
    model.fit(X_train, y_train)
    return model

def plot_confusion_matrix(conf_matrix):
    fig, ax = plt.subplots()
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Class 0', 'Class 1'],
                yticklabels=['Class 0', 'Class 1'], ax=ax)
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('True Label')
    ax.set_title('Confusion Matrix')

    buf = BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)

    img_str = base64.b64encode(buf.getvalue()).decode('utf-8')
    plt.close(fig)
    return img_str

app = dash.Dash(__name__)


model = train_model(X_train, y_train)

app.layout = html.Div([
    html.H1('Decision Tree Classifier'),

    dcc.Input(id='age', type='number', placeholder='Age'),
    dcc.Checklist(id='gender_male', options=[{'label': 'Gender: Male', 'value': 1}]),
    dcc.Checklist(id='sudden_weight_loss', options=[{'label': 'Sudden Weight Loss', 'value': 1}]),
    dcc.Checklist(id='weakness', options=[{'label': 'Weakness', 'value': 1}]),
    dcc.Checklist(id='polyphagia', options=[{'label': 'Polyphagia', 'value': 1}]),
    dcc.Checklist(id='genital_thrush', options=[{'label': 'Genital Thrush', 'value': 1}]),
    dcc.Checklist(id='irritability', options=[{'label': 'Irritability', 'value': 1}]),
    dcc.Checklist(id='delayed_healing', options=[{'label': 'Delayed Healing', 'value': 1}]),
    dcc.Checklist(id='partial_paresis', options=[{'label': 'Partial Paresis', 'value': 1}]),
    dcc.Checklist(id='muscle_stiffness', options=[{'label': 'Muscle Stiffness', 'value': 1}]),
    dcc.Checklist(id='alopecia', options=[{'label': 'Alopecia', 'value': 1}]),
    dcc.Checklist(id='obesity', options=[{'label': 'Obesity', 'value': 1}]),

    html.Button('Predict', id='predict-button', n_clicks=0),

    html.Div(id='prediction-output'),
    html.Div(id='accuracy-output'),
    html.Img(id='conf-matrix-plot')
])

@app.callback(
    [Output('prediction-output', 'children'),
     Output('accuracy-output', 'children'),
     Output('conf-matrix-plot', 'src')],
    [Input('predict-button', 'n_clicks')],
    [Input('age', 'value'),
     Input('gender_male', 'value'),
     Input('sudden_weight_loss', 'value'),
     Input('weakness', 'value'),
     Input('polyphagia', 'value'),
     Input('genital_thrush', 'value'),
     Input('irritability', 'value'),
     Input('delayed_healing', 'value'),
     Input('partial_paresis', 'value'),
     Input('muscle_stiffness', 'value'),
     Input('alopecia', 'value'),
     Input('obesity', 'value')]
)
def update_output(n_clicks, age, gender_male,  sudden_weight_loss, weakness,
                   polyphagia, genital_thrush, irritability,
                   delayed_healing, partial_paresis, muscle_stiffness, alopecia, obesity):

    if n_clicks > 0:
        input_data = np.array([[
            age if age is not None else 0,
            1 if gender_male else 0,
            1 if sudden_weight_loss else 0,
            1 if weakness else 0,
            1 if polyphagia else 0,
            1 if genital_thrush else 0,
            1 if irritability else 0,
            1 if delayed_healing else 0,
            1 if partial_paresis else 0,
            1 if muscle_stiffness else 0,
            1 if alopecia else 0,
            1 if obesity else 0
        ]])

        prediction = model.predict(input_data)[0]

        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        conf_matrix = confusion_matrix(y_test, y_pred)

        conf_matrix_plot = plot_confusion_matrix(conf_matrix)

        return (f'Prediction: {"Positive" if prediction == 1 else "Negative"}',
                f'Accuracy: {accuracy:.2f}',
                f'data:image/png;base64,{conf_matrix_plot}')

    return "", "", ""

if __name__ == '__main__':
    app.run_server(debug=True)
```
